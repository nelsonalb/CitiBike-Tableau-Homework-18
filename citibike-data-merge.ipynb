```python
import pandas as pd
```

```python
# Number of Months
n = 12


citibike_full_data = pd.DataFrame()

# For loop to read each month's CSV and add month's data to full_dataframe

for x in range(1, n+1):
    
    if x < 10:
        month = f'0{x}'
    else:
        month = x
    
    df = pd.read_csv(f'citibike-data/2022/JC-2022{month}-citibike-tripdata.csv')
    
    citibike_full_data = pd.concat([citibike_full_data, df])

citibike_full_data
```

```python
citibike_full_data.to_csv("citibike-data/citibike-2022.csv")
```
